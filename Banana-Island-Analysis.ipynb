```python
Avg_monthly_revenueA = (Property_A['Daily_Revenue'].mean() * Property_A['Occupancy_Days'].mean()) - (Property_A['Monthly_Cost']).mean()
Avg_monthly_revenueA.round(1)
```

```python
Avg_monthly_revenueB = (Property_B['Daily_Revenue'].mean() * Property_B['Occupancy_Days'].mean()) - (Property_B['Monthly_Cost']).mean()
Avg_monthly_revenueB.round(1)
```

```python
Avg_monthly_revenueC = (Property_C['Daily_Revenue'].mean() * Property_C['Occupancy_Days'].mean()) - (Property_C['Monthly_Cost']).mean()
Avg_monthly_revenueC.round(1)
```

```python
Yearly_revenueA = Avg_monthly_revenueA * 12
print(Yearly_revenueA.round(1))
```

```python
Yearly_revenueB = Avg_monthly_revenueB * 12
print(Yearly_revenueB.round(1))
```

```python
Yearly_revenueC = Avg_monthly_revenueC * 12
print(Yearly_revenueC.round(1))
```

```python
Best_Property_A_daily_income= Property_A['Daily_Revenue'].mean() + Property_A['Daily_Revenue'].std()
                            
Best_Property_A_Monthly_Cost= Property_A['Monthly_Cost'].mean() + Property_A['Monthly_Cost'].std()

Best_Property_A_occupancy= Property_A['Occupancy_Days'].mean() + Property_A['Occupancy_Days'].std()
                            
print(Best_Property_A_daily_income)    
print(Best_Property_A_Monthly_Cost)                           
print(Best_Property_A_occupancy)                           
```

```python
Best_Property_B_daily_income= Property_B['Daily_Revenue'].mean() + Property_B['Daily_Revenue'].std()
                            
Best_Property_B_Monthly_Cost= Property_B['Monthly_Cost'].mean() + Property_B['Monthly_Cost'].std()

Best_Property_B_occupancy= Property_A['Occupancy_Days'].mean() + Property_B['Occupancy_Days'].std()
                            
print(Best_Property_B_daily_income)    
print(Best_Property_B_Monthly_Cost)                           
print(Best_Property_B_occupancy)                           
```

```python
Best_Property_C_daily_income= Property_A['Daily_Revenue'].mean() + Property_C['Daily_Revenue'].std()
                            
Best_Property_C_Monthly_Cost= Property_A['Monthly_Cost'].mean() + Property_C['Monthly_Cost'].std()

Best_Property_C_occupancy= Property_A['Occupancy_Days'].mean() + Property_C['Occupancy_Days'].std()
                            
print(Best_Property_C_daily_income)    
print(Best_Property_C_Monthly_Cost)                           
print(Best_Property_C_occupancy)                           
```

```python
monthly_bestA = (Best_Property_A_daily_income * Best_Property_A_occupancy) - Best_Property_A_Monthly_Cost 
Yearly_bestA = monthly_bestA * 12
print(Yearly_bestA.round(1))
```

```python
monthly_bestB = (Best_Property_B_daily_income * Best_Property_B_occupancy) - Best_Property_B_Monthly_Cost 
Yearly_bestB = monthly_bestB * 12
print(Yearly_bestB.round(1))
```

```python
monthly_bestC = (Best_Property_C_daily_income * Best_Property_C_occupancy) - Best_Property_C_Monthly_Cost 
Yearly_bestC = monthly_bestC * 12
print(Yearly_bestC.round(1))
```

```python
worse_Property_A_daily_income= Property_A['Daily_Revenue'].mean() - Property_A['Daily_Revenue'].std()
                            
worse_Property_A_Monthly_Cost= Property_A['Monthly_Cost'].mean() - Property_A['Monthly_Cost'].std()

worse_Property_A_occupancy= Property_A['Occupancy_Days'].mean() - Property_A['Occupancy_Days'].std()
                            
print(worse_Property_A_daily_income)    
print(worse_Property_A_Monthly_Cost)                           
print(worse_Property_A_occupancy)                           
```

```python
worse_Property_B_daily_income= Property_B['Daily_Revenue'].mean() - Property_B['Daily_Revenue'].std()
                            
worse_Property_B_Monthly_Cost= Property_B['Monthly_Cost'].mean() - Property_B['Monthly_Cost'].std()

worse_Property_B_occupancy= Property_B['Occupancy_Days'].mean() - Property_B['Occupancy_Days'].std()
                            
print(worse_Property_B_daily_income)    
print(worse_Property_B_Monthly_Cost)                           
print(worse_Property_B_occupancy)                           
```

```python
worse_Property_C_daily_income= Property_C['Daily_Revenue'].mean() - Property_C['Daily_Revenue'].std()
                            
worse_Property_C_Monthly_Cost= Property_C['Monthly_Cost'].mean() - Property_C['Monthly_Cost'].std()

worse_Property_C_occupancy= Property_C['Occupancy_Days'].mean() - Property_C['Occupancy_Days'].std()
                            
print(worse_Property_C_daily_income)    
print(worse_Property_C_Monthly_Cost)                           
print(worse_Property_C_occupancy)                           
```

```python
monthly_worseA = (worse_Property_A_daily_income * worse_Property_A_occupancy) - worse_Property_A_Monthly_Cost 
Yearly_worseA = monthly_worseA * 12
Yearly_worseA.round(1)
```

I made charts to show the trend of each property's occupancy days, daily revenue and monthly cost over time 

```python
plt.figure(figsize=(10, 6))
plt.plot(Property_A['Month'], Property_A['Occupancy_Days'], marker='o', label='Property A', color='turquoise')
plt.plot(Property_B['Month'], Property_B['Occupancy_Days'], marker='*', label='property B', color='indigo')
plt.plot(Property_C['Month'], Property_C['Occupancy_Days'], marker='o', label='property C', color= 'k')
plt.xlabel('Months')
plt.ylabel('Occupancy days')
plt.title('Occupancy Days Over Time...All Properties')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Property_A['Month'], Property_A['Daily_Revenue'], marker='o', label='Property A', color='salmon')
plt.plot(Property_B['Month'], Property_B['Daily_Revenue'], marker='*', label='property B', color='green')
plt.plot(Property_C['Month'], Property_C['Daily_Revenue'], marker='s', label='property C', color= 'teal')
plt.xlabel('months')
plt.ylabel('daily revenue')
plt.title('daily revenue over time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(Property_A['Month'], Property_A['Monthly_Cost'], marker='o', label='Property A', color='blue')
plt.plot(Property_B['Month'], Property_B['Monthly_Cost'], marker='*', label='property B', color='plum')
plt.plot(Property_C['Month'], Property_C['Monthly_Cost'], marker='s', label='property C', color= 'hotpink')
plt.xlabel('months')
plt.ylabel('montly cost')
plt.title('monthly cost over time.... over time')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(Property_A['Occupancy_Days'], Property_A['Monthly_Cost'], color='hotpink', label='property A')
plt.scatter(Property_B['Occupancy_Days'], Property_B['Monthly_Cost'], color='indigo', label='property B')
plt.scatter(Property_C['Occupancy_Days'], Property_C['Monthly_Cost'], color='limegreen', label='property C')
plt.xlabel('Occupancy Days')
plt.ylabel('Monthly Cost')
plt.title('Check if daily occupancy affects the monthly cost')
plt.legend()
plt.show()
```
